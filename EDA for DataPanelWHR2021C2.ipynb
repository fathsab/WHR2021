```python
from sklearn.impute import SimpleImputer
import numpy as np

numerical_columns = dfs.select_dtypes(include=[np.number]).columns

# transform numerical columns into NumPy arrays
numerical_arrays = dfs[numerical_columns].to_numpy()

# fill missing values with mean using scikit-learn
imputer = SimpleImputer(strategy='mean')
numerical_arrays_filled = imputer.fit_transform(numerical_arrays)

# transform filled NumPy arrays back into dataframe columns
dfs[numerical_columns] = numerical_arrays_filled
```

```python
# check for missing values
missing_values = dfs.isnull()

# count missing values in each column
missing_values_count = missing_values.sum()

print(missing_values_count)
```

```python
import matplotlib.pyplot as plt

data = dfs[["Country name", "year", "Life Ladder"]]

# Pivot the data to have each country as a separate column
data = data.pivot(index="year", columns="Country name", values="Life Ladder")

# Create a line chart of the life ladder scores over time for each country
data.plot(kind="line", figsize=(12, 8))

# Create a bar chart of the average life ladder score for each country over the entire time period
data.mean().plot(kind="bar", figsize=(12, 8))

# Set the chart title and axis labels
plt.title("Life Ladder Score by Country and Year")
plt.xlabel("Year")
plt.ylabel("Life Ladder Score")

# Show the chart
plt.show()
```
